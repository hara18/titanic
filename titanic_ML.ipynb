```python
full_df = full_df.drop(columns = ["Cabin", "Ticket", "Name", "Embarked", "PassengerId"], axis = 1) #axis 1 = x, 0 = y
```

```python
#заполнение нулевых значений с помощью медианы
full_df["Age"] = full_df["Age"].fillna(full_df["Age"].median())
full_df["Survived"] = full_df["Survived"].fillna(full_df["Survived"].median())
full_df["Fare"] = full_df["Fare"].fillna(full_df["Fare"].median())
```

```python
#импортирование класса для кодирования столбца с полом в двоичный формат
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
full_df["Sex"] = le.fit_transform(full_df["Sex"])
```

```python
full_df["Age"] = full_df["Age"].astype("int64")
full_df["Survived"] = full_df["Survived"].astype("int64")
full_df["Sex"] = full_df["Sex"].astype("int64")
```

```python
#ML
```

```python
#разделение на x и y (отдельные признаки и целевая переменная)
X = full_df.drop(columns = "Survived", axis = 1)
y = full_df["Survived"].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, test_size = 0.2)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
#иморт необходимого класса, метод соседей
from sklearn.neighbors import KNeighborsClassifier
#создаём три списка в которых будет лежать кол-во соседей, а остальные результат тренировочной и тестовой
neighbors = [3,6,9]
train_score =[]
test_score=[]
#запускаем цикл
for neigh in neighbors:
    kn = KNeighborsClassifier(n_jobs = -1, n_neighbors = neigh)
    kn.fit(X_train, y_train)
    train_score.append(kn.score(X_train, y_train))
    test_score.append(kn.score(X_test, y_test))
```

```python
print("Train", train_score)
print("Test", test_score)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, kn.predict(X_test)))
```

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
```

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
forest = RandomForestClassifier()
forest.fit(X_train, y_train)
```

```python
forest.score(X_test, y_test)
```

```python
grad = GradientBoostingClassifier()
grad.fit(X_train, y_train)
```

```python
grad.score(X_test, y_test)
```

```python
print("Метод деревьев",classification_report(y_test, forest.predict(X_test)))
print("Метод увеличения градиента",classification_report(y_test, grad.predict(X_test)))
print("Метод ближайших соседей",classification_report(y_test, kn.predict(X_test)))
```
